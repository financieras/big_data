# Lección 1.3.2: Excel: Estructura de Datos y Funciones

## 1. El Rol de Excel: El Punto de Partida del Análisis

En el ecosistema de datos moderno (donde se usan Python, SQL y la Nube), **Excel sigue siendo la herramienta más usada en un primer análisis**.

-   **¿Por qué?** Es rápido, accesible y permite un análisis *ad-hoc* (sobre la marcha) sin necesidad de programar.
-   **Nuestro Enfoque:** Usaremos Excel para la **preparación y transformación de datos (el primer ETL)**, sentando las bases antes de pasar a herramientas más potentes.

> **Analogía:** Excel es la **navaja suiza** del analista; imprescindible para tareas rápidas y específicas.

---

## 2. La Estructura es el Fundamento: Preparando Datos "Tidy"

El mayor problema de Excel es que permite *demasiada* libertad. Para que un dato sea útil para el análisis, debe ser **estructurado**.

### 2.1. La Regla de Oro del "Tidy Data" (Datos Ordenados)

Para que el análisis funcione (en Excel, Python o SQL), los datos deben seguir tres reglas de oro:

| Elemento | Descripción | Ejemplo de error |
| :--- | :--- | :--- |
| **Variable** | Cada variable (ej. `Fecha`, `Región`, `Ventas`) debe tener su **propia columna**. | Tener la región en el encabezado de la columna. |
| **Observación** | Cada observación (ej. una transacción, un cliente) debe tener su **propia fila**. | Usar varias filas para el mismo registro. |
| **Valor** | Cada valor individual debe ocupar una **sola celda**. | Celdas combinadas o múltiples valores en una sola celda. |

### 2.2. Tablas Estructuradas: ¡Dile adiós a los rangos!

Un **Rango de Datos** es solo un área de celdas. Una **Tabla Estructurada** (insertada con **CTRL + T** o desde la pestaña "Insertar") ofrece ventajas clave:

-   **Rangos Dinámicos:** La tabla crece automáticamente al añadir filas.
-   **Nombres:** Puedes referenciar la tabla por un nombre (`TableSales[Sales]`) en lugar de por celdas (`A1:C500`).
-   **Fórmulas Inteligentes:** Las fórmulas se copian automáticamente en toda la columna.

## 3. Fórmulas Clave para la Transformación (Limpieza de Datos)

Las funciones en Excel son la forma de transformar y manipular los datos. Son nuestro primer **lenguaje de programación** para la limpieza.

### 3.1. Funciones de Búsqueda: El "JOIN" Básico (`XLOOKUP`)

Sirven para traer datos de una tabla a otra, buscando una coincidencia.

| Función | Uso | Analista lo ve como... |
| :--- | :--- | :--- |
| **`XLOOKUP`** (Recomendada) | Busca un valor en una columna y devuelve el valor de otra columna relacionada. | **Unir** dos tablas usando una clave común (`JOIN`). |
| **`VLOOKUP`** (Versión Antigua) | Solo busca a la derecha (menos flexible). | Un `LEFT JOIN` limitado. |

> **Caso de Uso:** Tienes el `ID_Cliente` en tu tabla de ventas y quieres añadir la columna `Nombre_Cliente` desde tu tabla maestra de clientes. ¡Usa `XLOOKUP`!

### 3.2. Funciones Lógicas: Creación de Variables y Control de Errores

| Función | Propósito | Ejemplo Práctico |
| :--- | :--- | :--- |
| **`IF`** | Evalúa una condición y devuelve un valor si es verdadera y otro si es falsa. | Clasificar `IF(Sales > 1000, "Premium", "Standard")`. |
| **`AND` / `OR`** | Combina múltiples condiciones dentro de un `IF`. | `IF(AND(Country="MX", Amount>500), "Alert", "Normal")`. |
| **`IFERROR`** | Muestra un valor alternativo cuando una fórmula genera un error (ej. al no encontrar un `XLOOKUP`). | `IFERROR(XLOOKUP(ID),"No Encontrado")`. |

### 3.3. Funciones de Texto: Higiene del Dato

| Función | Propósito |
| :--- | :--- |
| **`TRIM`** | Elimina espacios extra al inicio, al final y espacios duplicados dentro de la cadena (¡Súper importante!). |
| **`UPPER` / `LOWER`** | Homogeneiza el formato de mayúsculas/minúsculas para evitar problemas de conteo (ej. "Mexico" vs. "mexico"). |
| **`CONCAT`** | Une el contenido de varias celdas o cadenas de texto. |
| **`MID`** / **`LEFT`** / **`RIGHT`** | Extrae un número específico de caracteres de una cadena. |

## 4. Agregación Condicional: El Primer "GROUP BY"

Las funciones condicionales permiten resumir los datos (sumar, contar, promediar) **solo si cumplen un criterio**. Esto es lo que se conoce en bases de datos como **`GROUP BY`**.

-   **`SUMIFS`:** Suma los valores de un rango si se cumplen **uno o más criterios**.
-   **`AVERAGEIFS`:** Calcula la media de un rango si se cumplen **uno o más criterios**.
-   **`COUNTIFS`:** Cuenta las celdas de un rango si se cumplen **uno o más criterios**.

> **Caso Práctico:** ¿Cuál es la **suma de ventas** (`SUMIFS`) en la **Región Norte** (`Criterio 1`) para el **Producto A** (`Criterio 2`)?

## 5. Resumen

| Tarea | Rol de la Función | Función Clave (EN) |
| :--- | :--- | :--- |
| **Estructura** | Garantizar un análisis limpio. | **CTRL + T** (Structured Table) |
| **Relación** | Unir datos de diferentes fuentes. | **`XLOOKUP`** (o `VLOOKUP`) |
| **Lógica** | Crear variables o manejar errores. | **`IF`** y **`IFERROR`** |
| **Higiene** | Limpiar formatos de texto. | **`TRIM`** y **`UPPER`** |
| **Agregación** | Resumir datos por categorías. | **`SUMIFS`** |

> **El Takeaway:** Antes de avanzar a herramientas más complejas, dominar la estructura y las funciones de Excel te permite validar y limpiar datos, ahorrando horas de trabajo posteriores.
