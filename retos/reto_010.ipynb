# **Reto 10. VisualizaciÃ³n avanzada con boxplot y violinplot**
- Crea boxplot de sepal_width por especie
- Crea violinplot de la misma variable
- Identifica especie con mayor mediana de sepal_width
- Identifica especie con mayor variabilidad

# **ResoluciÃ³n del Reto 10**

## **Paso 1: Cargar dataset Iris desde Seaborn**
Importamos el dataset clÃ¡sico de flores Iris para el anÃ¡lisis de visualizaciÃ³n.

```python
# EJECUTA ESTE CÃ“DIGO PARA CARGAR IRIS
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("ðŸŒ¸ Paso 1: Cargando dataset Iris...")
df = sns.load_dataset('iris')
print(f"âœ… Dataset cargado - Dimensiones: {df.shape}")
print(f"ðŸŒº Especies disponibles: {df['species'].unique()}")
```

## **Paso 2: Crear boxplot de sepal_width por especie**
Visualizamos la distribuciÃ³n del ancho del sÃ©palo para cada especie usando boxplot.

```python
# EJECUTA ESTE CÃ“DIGO PARA CREAR BOXPLOT
print("ðŸ“¦ Paso 2: Creando boxplot de sepal_width por especie...")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal_width', palette='Set2')
plt.title('DistribuciÃ³n del Ancho del SÃ©palo por Especie (Boxplot)')
plt.xlabel('Especie')
plt.ylabel('Ancho del SÃ©palo (cm)')
plt.grid(axis='y', alpha=0.3)
plt.show()
print("âœ… Boxplot creado correctamente")
```

## **Paso 3: Crear violinplot de sepal_width por especie**
Complementamos con violinplot para visualizar la densidad de la distribuciÃ³n.


```python
# EJECUTA ESTE CÃ“DIGO PARA CREAR VIOLINPLOT
print("ðŸŽ» Paso 3: Creando violinplot de sepal_width por especie...")
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='species', y='sepal_width', palette='Set2')
plt.title('DistribuciÃ³n del Ancho del SÃ©palo por Especie (Violinplot)')
plt.xlabel('Especie')
plt.ylabel('Ancho del SÃ©palo (cm)')
plt.grid(axis='y', alpha=0.3)
plt.show()
print("âœ… Violinplot creado correctamente")
```

## **Paso 4: Identificar especie con mayor mediana de sepal_width**
Calculamos las medianas para determinar quÃ© especie tiene el sÃ©palo mÃ¡s ancho.

```python
# EJECUTA ESTE CÃ“DIGO PARA CALCULAR MEDIANAS
print("ðŸ“Š Paso 4: Calculando medianas por especie...")
medianas = df.groupby('species')['sepal_width'].median().sort_values(ascending=False)
print("Medianas de sepal_width por especie:")
print(medianas)

especie_mayor_mediana = medianas.index[0]
print(f"\nðŸŽ¯ Especie con mayor mediana de sepal_width: {especie_mayor_mediana}")
print(f"ðŸ“ Mediana: {medianas.iloc[0]:.2f} cm")
```

## **Paso 5: Identificar especie con mayor variabilidad**
Calculamos la desviaciÃ³n estÃ¡ndar para determinar quÃ© especie presenta mayor variabilidad.

```python
# EJECUTA ESTE CÃ“DIGO PARA CALCULAR VARIABILIDAD
print("ðŸ“ˆ Paso 5: Calculando variabilidad por especie...")
desviaciones = df.groupby('species')['sepal_width'].std().sort_values(ascending=False)
print("DesviaciÃ³n estÃ¡ndar de sepal_width por especie:")
print(desviaciones)

especie_mayor_variabilidad = desviaciones.index[0]
print(f"\nðŸŽ¯ Especie con mayor variabilidad en sepal_width: {especie_mayor_variabilidad}")
print(f"ðŸ“ DesviaciÃ³n estÃ¡ndar: {desviaciones.iloc[0]:.3f} cm")
```

## **Paso 6: VisualizaciÃ³n combinada (opcional)**
Creamos una visualizaciÃ³n que combina boxplot y violinplot para comparaciÃ³n.

```python
# EJECUTA ESTE CÃ“DIGO PARA VISUALIZACIÃ“N COMBINADA
print("ðŸ”„ Paso 6: Creando visualizaciÃ³n combinada...")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Boxplot
sns.boxplot(data=df, x='species', y='sepal_width', palette='Set2', ax=ax1)
ax1.set_title('Boxplot - Ancho del SÃ©palo por Especie')
ax1.set_xlabel('Especie')
ax1.set_ylabel('Ancho del SÃ©palo (cm)')
ax1.grid(axis='y', alpha=0.3)

# Violinplot
sns.violinplot(data=df, x='species', y='sepal_width', palette='Set2', ax=ax2)
ax2.set_title('Violinplot - Ancho del SÃ©palo por Especie')
ax2.set_xlabel('Especie')
ax2.set_ylabel('Ancho del SÃ©palo (cm)')
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
print("âœ… VisualizaciÃ³n combinada creada")
```

## **Paso 7: Resumen de hallazgos**
Sintetizamos los principales descubrimientos del anÃ¡lisis.

```python
# EJECUTA ESTE CÃ“DIGO PARA RESUMEN
print("ðŸŽ“ Paso 7: Resumen de hallazgos del Reto 10")
print("=" * 50)
print(f"ðŸ“Š Dataset Iris: {df.shape[0]} observaciones, {df.shape[1]} caracterÃ­sticas")
print(f"ðŸŒº Especies analizadas: {', '.join(df['species'].unique())}")
print(f"ðŸŽ¯ Especie con MAYOR mediana de sepal_width: {especie_mayor_mediana}")
print(f"ðŸŽ¯ Especie con MAYOR variabilidad en sepal_width: {especie_mayor_variabilidad}")
print("\nðŸ“‹ InterpretaciÃ³n visual:")
print("   â€¢ Boxplot: Muestra medianas, cuartiles y valores atÃ­picos")
print("   â€¢ Violinplot: Combina boxplot con densidad de probabilidad")
print("   â€¢ Setosa: SÃ©palo mÃ¡s ancho y menos variable")
print("   â€¢ Virginica: Mayor variabilidad en el ancho del sÃ©palo")

print("\nðŸŽ‰ Â¡Reto 10 completado! Dominada la visualizaciÃ³n avanzada con boxplot y violinplot")
```
