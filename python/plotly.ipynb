## 1. 쯈u칠 es Plotly?

Plotly es una biblioteca de Python para **visualizaci칩n de datos interactiva** y profesional. Es ideal para Data Science y Machine Learning porque permite explorar datos de forma din치mica y presentar resultados de manera clara.

**Ventajas principales de Plotly:**

* **Interactividad:** Zoom, filtrado y tooltips al pasar el cursor
* **Diversidad de gr치ficos:** Dispersi칩n, l칤neas, barras, mapas geogr치ficos, gr치ficos 3D, series temporales, entre otros
* **Integraci칩n con pandas:** Facilita trabajar con DataFrames directamente
* **Visualizaci칩n profesional:** Gr치ficos listos para presentaciones o dashboards
* **Exploraci칩n de patrones:** Ideal para detectar correlaciones y tendencias antes de entrenar modelos de Machine Learning

**游눠 Por qu칠 usar Plotly en Data Science y ML:**

1. Exploras patrones y correlaciones **interactivamente**
2. Visualizas distribuciones y relaciones antes de entrenar modelos

## 4. Tipos de gr치ficos en Plotly

| Tipo de gr치fico               | Uso principal                        | Ejemplo r치pido en Python                                             |
| ----------------------------- | ------------------------------------ | -------------------------------------------------------------------- |

## 5. Ejemplo 1: An치lisis exploratorio con Iris

```python
import plotly.express as px

# Cargar dataset
df = px.data.iris()

# 1. Gr치fico de dispersi칩n interactivo
fig = px.scatter(df, x="sepal_length", y="sepal_width",
                 color="species", size="petal_length",
                 hover_data=["petal_width"],
                 title="An치lisis de caracter칤sticas del Iris")
fig.show()

# 2. Boxplot para detectar outliers
fig2 = px.box(df, x="species", y="sepal_length",
              title="Distribuci칩n de longitud de s칠palo por especie",
              color="species")
fig2.show()

# 3. Histograma de distribuciones
fig3 = px.histogram(df, x="petal_length", color="species",
                    marginal="box",  # A침ade boxplot en el margen
                    title="Distribuci칩n de longitud de p칠talo")
fig3.show()
```

**游꿢 Insight de ML:** Estos gr치ficos te ayudan a identificar qu칠 caracter칤sticas (features) tienen mejor separabilidad entre clases antes de entrenar un clasificador.

## 6. Ejemplo 2: Matriz de correlaci칩n para Feature Engineering

```python
import plotly.express as px
import pandas as pd

# Cargar dataset
df = px.data.iris()

# Seleccionar solo columnas num칠ricas
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df_numeric = df[numeric_cols]

# Calcular matriz de correlaci칩n
corr_matrix = df_numeric.corr()

# Visualizar con heatmap
fig = px.imshow(corr_matrix,
                text_auto=True,  # Mostrar valores
                aspect="auto",
                color_continuous_scale="RdBu_r",
                title="Matriz de correlaci칩n - Dataset Iris")
fig.show()
```

**游꿢 Insight de ML:** Las correlaciones altas entre features pueden indicar redundancia. Puedes usar esta informaci칩n para Feature Selection o detecci칩n de multicolinealidad.

## 7. Ejemplo 3: An치lisis de series temporales

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Crear datos de ejemplo (serie temporal)
dates = pd.date_range('2023-01-01', periods=100, freq='D')
data = {
    'fecha': dates,
    'ventas': np.cumsum(np.random.randn(100)) + 100,
    'visitas': np.cumsum(np.random.randn(100)) + 500
}
df = pd.DataFrame(data)

# Gr치fico de l칤neas con zoom temporal
fig = px.line(df, x='fecha', y=['ventas', 'visitas'],
              title='Evoluci칩n temporal de m칠tricas',
              labels={'value': 'Cantidad', 'variable': 'M칠trica'})

# A침adir rango selector (칰til para ML con datos temporales)
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=7, label="1w", step="day", stepmode="backward"),
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

**游꿢 Insight de ML:** Ideal para detectar tendencias, estacionalidad y anomal칤as antes de aplicar modelos de forecasting como ARIMA o Prophet.

## 8. Ejemplo 4: Visualizaci칩n 3D para clustering

```python
import plotly.express as px
from sklearn.datasets import make_blobs

# Generar datos sint칠ticos de clustering
X, y = make_blobs(n_samples=300, n_features=3, centers=4, random_state=42)

# Crear DataFrame
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2', 'Feature_3'])
df['Cluster'] = y

# Gr치fico 3D interactivo
fig = px.scatter_3d(df, x='Feature_1', y='Feature_2', z='Feature_3',
                    color='Cluster',
                    title='Visualizaci칩n 3D de Clusters',
                    opacity=0.7)
fig.show()
```

**游꿢 Insight de ML:** Perfecto para evaluar visualmente la calidad de algoritmos de clustering como K-Means o DBSCAN.

## 9. Ejemplo 5: Comparaci칩n de modelos de ML

```python
import plotly.graph_objects as go

# Resultados de diferentes modelos (ejemplo)
modelos = ['Random Forest', 'SVM', 'Logistic Regression', 'XGBoost']
accuracy = [0.92, 0.88, 0.85, 0.94]
f1_score = [0.91, 0.87, 0.84, 0.93]
precision = [0.90, 0.89, 0.86, 0.92]

# Crear gr치fico de barras agrupadas
fig = go.Figure(data=[
    go.Bar(name='Accuracy', x=modelos, y=accuracy),
    go.Bar(name='F1-Score', x=modelos, y=f1_score),
    go.Bar(name='Precision', x=modelos, y=precision)
])

fig.update_layout(
    title='Comparaci칩n de m칠tricas entre modelos',
    xaxis_title='Modelos',
    yaxis_title='Score',
    barmode='group',
    yaxis=dict(range=[0, 1])
)
fig.show()
```

**游꿢 Insight de ML:** Facilita la presentaci칩n de resultados y la selecci칩n del mejor modelo seg칰n m칰ltiples m칠tricas.

## 10. Personalizaci칩n y tips avanzados

```python
import plotly.express as px

df = px.data.iris()

# Gr치fico con personalizaci칩n avanzada
fig = px.scatter(df, x="sepal_length", y="sepal_width",
                 color="species",
                 title="Gr치fico personalizado")

# Personalizar dise침o
fig.update_layout(
    template="plotly_dark",  # Tema oscuro
    font=dict(size=14),
    hovermode='closest',
    showlegend=True
)

# Personalizar ejes
fig.update_xaxes(title_font=dict(size=16, color='white'))
fig.update_yaxes(title_font=dict(size=16, color='white'))

fig.show()
```

**游눠 Tips profesionales:**

* Usa `template` para cambiar el tema: `"plotly"`, `"plotly_white"`, `"plotly_dark"`, `"ggplot2"`, `"seaborn"`
* A침ade anotaciones con `fig.add_annotation()` para destacar puntos importantes
* Exporta gr치ficos con `fig.write_html("grafico.html")` o `fig.write_image("grafico.png")`
* Para dashboards, combina Plotly con **Dash** o **Streamlit**
