## 2\. SQL: El Mundo Relacional (Estructura y Tablas)

### 2.1. El Contrato de Consistencia (ACID)

El gran valor de SQL es la **integridad de los datos**. Esto se consigue con las propiedades **ACID** (Atomicidad, Consistencia, Aislamiento y Durabilidad), que act칰an como un contrato de fiabilidad para cada transacci칩n:

| Propiedad | Descripci칩n | Ejemplo |
| :--- | :--- | :--- |

| **Aislamiento** | Las transacciones que ocurren al mismo tiempo no se interfieren. | Dos usuarios comprando el 칰ltimo billete de avi칩n: solo uno puede obtenerlo. |
| **Durabilidad** | Una vez que se confirma la transacci칩n, el cambio es permanente, incluso si el sistema falla. | Una vez que guardas un registro, la DB lo garantiza. |

## 3\. NoSQL: La Flexibilidad para Big Data (Schema-less)

El auge de las bases de datos NoSQL se debe a dos grandes retos que SQL no resolv칤a bien: la **escalabilidad horizontal** (distribuir la carga en muchos servidores) y la necesidad de manejar datos **semi-estructurados** o que cambian muy r치pido.

### 3.1. Tipos de Bases NoSQL y Flexibilidad

Las bases NoSQL (como MongoDB o Cassandra) sacrifican la estricta consistencia ACID por una mayor **disponibilidad** y **flexibilidad**.

  - **Sin Esquema Fijo (*Schema-less*):** Puedes almacenar un registro con 5 campos y el siguiente con 15, ideal para datos de *logs* de servidores, redes sociales o IoT.
  - **Tipos Comunes:**
      - **Documentales:** Almacenan documentos (ej. MongoDB).
      - **Clave-Valor:** Almacenan pares simples (ej. Redis).
      - **Orientadas a Grafos:** Para datos con relaciones complejas (ej. Neo4j).

### 3.2. JSON: El Lenguaje de los Documentos NoSQL

El formato predilecto para almacenar datos en las bases NoSQL Documentales es **JSON** (*JavaScript Object Notation*). Es un formato legible que almacena la informaci칩n como pares **clave-valor** anidados.

**Ejemplo de un Documento JSON de Producto (MongoDB):**

```json
{
  "_id": "PROD_45",
  "nombre": "Laptop Ultraligera",
  "precio": 999.99,
  "proveedor": "TechCorp",
  "especificaciones": {
    "ram_gb": 16,
    "almacenamiento_gb": 512,
    "dimensiones_cm": "30x20x1.5"
  },
  "reviews_ids": [123, 124, 125]
}
```

> **Observa:** Toda la informaci칩n del producto, incluidas sus especificaciones, est치 **embe침ida** (anidada) en un solo documento, eliminando la necesidad de m칰ltiples *JOINs*.

## 4\. SQL vs NoSQL: El Debate del Ecosistema

| Caracter칤stica | SQL (Relacional) | NoSQL (Documental/Clave-Valor) |
| :--- | :--- | :--- |
| **Estructura** | Estricta (esquema fijo), basada en **Tablas**. | Flexible (sin esquema fijo), basada en **Documentos/Objetos**. |
| **Consistencia** | Alta (garant칤a **ACID**). Prioriza la integridad. | Baja (**Consistencia Eventual**). Prioriza la velocidad y disponibilidad. |
| **Escalabilidad** | Vertical (mejorar un 칰nico servidor). | Horizontal (a침adir m치s servidores, *sharding*). |
| **Uso Ideal** | Sistemas transaccionales cr칤ticos (Banca, Contabilidad), datos estructurados. | Big Data, datos en tiempo real, cat치logos de productos, datos de usuario web. |

## 5\. Resumen y Conclusiones

1.  **SQL** se basa en las **relaciones** entre tablas para mantener la **integridad (ACID)** y es ideal para datos estructurados donde la exactitud es primordial (ej. dinero).
2.  **NoSQL** utiliza formatos flexibles como **JSON** para almacenar documentos completos, priorizando la **velocidad** y la **escalabilidad horizontal** para entornos de Big Data.
3.  El profesional de datos moderno no elige uno sobre el otro; utiliza el modelo **SQL** o **NoSQL** que mejor se adapte a las necesidades del dato y del negocio.
