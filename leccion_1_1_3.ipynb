# Lección 1.1.3: Roles: Analista de Datos, Científico de Datos, Ingeniero de Datos

## 1. ¿Por qué roles especializados?

En los inicios del Big Data, el "Data Scientist" lo hacía todo: programar, analizar, visualizar. Hoy, la complejidad exige **roles especializados** que trabajan como una orquesta:  

- **Data Analyst**: Explica el presente. Convierte datos en respuestas para el negocio.
- **Data Scientist**: Predice el futuro con modelos avanzados.
- **Data Engineer**: Construye los cimientos. Se encarga de la infraestructura para que todo funcione.

> **Clave:**  Juntos transforman datos crudos en decisiones inteligentes. Juntos crean valor.

---

## 2. Data Analyst: El traductor del negocio

**Misión:** Transformar datos en **insights accionables** para decisiones empresariales.  

**Responsabilidades:**  
- Responder preguntas de negocio con datos.  
- Crear dashboards y reportes (Tableau, Power BI).  
- Analizar tendencias y KPIs.  
- Comunicar resultados a equipos no técnicos.  

**Perfil:**  
- **Habilidades:** SQL, herramientas BI, estadística básica, narrativa.  
- **Enfoque:** "¿Qué pasó y por qué?"  

**Ejemplo:**  
Un analista en Spotify descubre que los usuarios abandonan podcasts largos en móviles. Crea un dashboard y recomienda formatos más cortos. **Impacto:** +10% en retención.  

---

## 3. Data Scientist: El predictor del futuro

**Misión:** Usar estadística y machine learning para **predecir y optimizar**.  

**Responsabilidades:**  
- Construir modelos predictivos (churn, recomendaciones).  
- Diseñar experimentos (A/B testing).  
- Descubrir patrones complejos.  
- Validar hipótesis con rigor científico.  

**Perfil:**    
- **Habilidades:** Python, machine learning, estadística avanzada.  
- **Enfoque:** "¿Qué pasará y qué hacemos?"  

**Ejemplo:**  
Un científico en Netflix entrena un modelo que usa historial de visualización y predice cancelaciones con 85% de precisión, identificando usuarios en riesgo. **Impacto:** Reduce churn en 3%, ahorrando millones.  

---

## 4. Data Engineer: El arquitecto invisible

**Misión:** Diseñar y mantener **pipelines de datos** confiables y escalables que garanticen el flujo de información.

**Responsabilidades:**  
- Diseñar y construir pipelines (ETL/ELT) para mover datos.  
- Optimizar bases de datos (Snowflake, BigQuery).  
- Garantizar calidad y disponibilidad.  
- Automatizar procesos.  

**Perfil:**    
- **Habilidades:** Python/Scala, Spark, Airflow, cloud (AWS, GCP).  
- **Enfoque:** "¿Cómo llegan los datos rápido y limpios?"  

**Ejemplo:**  
Un ingeniero en Uber crea un pipeline que procesa 100 TB de datos de viajes en tiempo real. **Impacto:** Analistas acceden a datos frescos en minutos, no en horas.  

> **Analogía**: Si los datos son agua, el ingeniero construye las tuberías.

## 5. Comparación rápida

| Aspecto | Data Analyst | Data Scientist | Data Engineer |
|---------|--------------|----------------|---------------|
| **Pregunta** | ¿Qué pasó? | ¿Qué pasará? | ¿Cómo llegan los datos? |
| **Output** | Dashboards, insights | Modelos, predicciones | Pipelines, datos limpios |
| **Herramienta clave** | SQL, Power BI | Python, Scikit-learn | Spark, Airflow |
| **Formación** | Grado general | Grado STEM + máster | Ingeniería |
| **Matemáticas** | Básica | Avanzada | Mínima |
| **Barrera de entrada** | Media | Alta | Alta |

---

## 6. Colaboración: El flujo de datos

**Ejemplo en Amazon:**  
1. **Ingeniero:** Construye pipeline que une datos de ventas y logística.  
2. **Analista:** Crea dashboard que muestra exceso de stock por región.  
3. **Científico:** Desarrolla modelo para predecir demanda.  
4. **Ingeniero:** Pone el modelo en producción.  
5. **Analista:** Monitoriza KPIs para medir impacto.  

**Resultado:** -15% en costes de inventario, +10% en disponibilidad.  

> **Clave:** Los ingenieros habilitan, los analistas clarifican, los científicos predicen.

---

## 7. ¿Cuál es tu camino?

- **Data Analyst:** Si te gusta contar historias con datos y resolver problemas de negocio.  
- **Data Scientist:** Si amas matemáticas, experimentos y programar modelos.  
- **Data Engineer:** Si disfrutas al construir sistemas robustos y escalables.  

> **Tip:** Muchos empiezan como analistas y luego se especializan.

## 9. Otros Roles en el Ecosistema Moderno de Datos
A medida que los equipos de datos crecen, emergen roles especializados para abordar desafíos específicos de escalabilidad, análisis avanzado y operación. Estos roles complementan a los tres pilares (Analista, Científico, Ingeniero) y permiten que el ecosistema de datos sea más eficiente y adaptable.



| Rol | Enfoque Principal | ¿Cuándo suele aparecer? | Ejemplo de Impacto |
|-----|-------------------|--------------------------|---------------------|
| **Business Intelligence (BI) Specialist** | Diseñar, desarrollar y mantener soluciones de reporting y dashboards para la toma de decisiones. Trabaja en la intersección entre datos y negocio. | En empresas con fuerte cultura de reporting o donde los stakeholders necesitan acceso rápido a métricas clave. | En una cadena de retail crea un dashboard que unifica datos de ventas, inventario y logística, reduciendo de 2 días a 2 horas la generación de informes. |
| **Machine Learning Engineer** | Llevar modelos de ML a producción, optimizando su rendimiento, escalabilidad y latencia. | Cuando los prototipos de los científicos de datos necesitan desplegarse en sistemas reales y ser consumidos por aplicaciones o usuarios finales. | Un MLE en un banco implementa un modelo de detección de fraude en tiempo real, reduciendo falsos positivos en un 20%. |
| **MLOps Engineer** | Automatizar el ciclo de vida completo del ML (CI/CD, monitorización, re-entrenamiento, gobernanza). | Cuando hay muchos modelos en producción que requieren mantenimiento, vigilancia constante y actualizaciones frecuentes. | Un MLOps Engineer en una plataforma de streaming automatiza el re-entrenamiento de modelos de recomendación, mejorando la personalización en un 15%. |
| **Data Architect** | Diseñar la arquitectura global de datos, seleccionar tecnologías y planificar la escalabilidad a largo plazo. | En organizaciones con sistemas de datos complejos, migraciones importantes (ej. a la nube) o necesidades de integración de múltiples fuentes. | Un Data Architect en una aseguradora diseña un data lake unificado que reduce la duplicidad de datos y acelera el acceso a información crítica para subscriptores. |
| **Analytics Engineer** | Transformar datos crudos en conjuntos limpios, confiables y documentados (usando SQL, dbt, Dataform), listos para el análisis. | Cuando los analistas necesitan datos de alta calidad y los ingenieros están saturados con tareas de infraestructura. | Un Analytics Engineer en una startup de SaaS estandariza las métricas de crecimiento, reduciendo discrepancias entre equipos en un 90%. |
| **Data Governance Specialist** | Garantizar la calidad, seguridad, privacidad y cumplimiento normativo de los datos (GDPR, CCPA). | En empresas con regulaciones estrictas o que manejan datos sensibles (salud, finanzas). | Un especialista en una clínica implementa políticas de anonimización de datos, permitiendo el uso de información médica para investigación sin violar HIPAA. |
| **Data Product Manager** | Definir la visión y hoja de ruta de productos basados en datos, alineando necesidades de negocio con capacidades técnicas. | Cuando los proyectos de datos requieren gestión ágil y priorización estratégica, especialmente en equipos multidisciplinares. | Un Data PM en una fintech lidera el desarrollo de un "score de riesgo crediticio" que aumenta la aprobación de préstamos en un 12% sin incrementar el riesgo. |
| **Embedded Analyst** | Analista integrado en un equipo de negocio (marketing, operaciones), enfocado en resolver problemas específicos de ese área. | En organizaciones que adoptan un modelo de "data mesh" o descentralizan el análisis. | Un Embedded Analyst en el equipo de marketing optimiza el gasto en publicidad digital, aumentando el ROI en un 25%. |
| **Data Storyteller** | Comunicar insights complejos de manera clara y persuasiva, usando visualizaciones y narrativas efectivas. | Cuando los stakeholders no técnicos necesitan entender el valor de los datos sin profundizar en detalles técnicos. | Un Data Storyteller en una ONG convierte datos de impacto social en historias visuales que aumentan las donaciones en un 30%. |

**Visión global:**
Estos roles no reemplazan a los fundamentales (Analista, Científico, Ingeniero), sino que **complementan el ecosistema**, permitiendo que los equipos sean más eficientes, los sistemas más resilientes y las decisiones más informadas a medida que la organización madura en su uso de datos.


---

## 10. Resumen

- **Analista:** Usa SQL y BI para entender el presente.  
- **Científico:** Usa ML para predecir el futuro.  
- **Ingeniero:** Usa cloud y pipelines para contruir el sistema que hace posible todo.  

Juntos convierten datos crudos en **decisiones inteligentes**, formando un equipo de datos moderno.
