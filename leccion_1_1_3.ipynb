# Lección 1.1.3: Roles: Analista de Datos, Científico de Datos, Ingeniero de Datos

## 1. ¿Por qué roles especializados?

En los inicios del Big Data, el "Data Scientist" lo hacía todo: programar, analizar, visualizar. Hoy, la complejidad exige **roles especializados** que trabajan como una orquesta:  

- **Data Analyst**: Explica el presente. Convierte datos en respuestas para el negocio.
- **Data Scientist**: Predice el futuro con modelos avanzados.
- **Data Engineer**: Construye los cimientos. Se encarga de la infraestructura para que todo funcione.

> **Clave:**  Juntos transforman datos crudos en decisiones inteligentes. Juntos crean valor.

## 2. Data Analyst: El traductor del negocio

**Misión:** Transformar datos en **insights accionables** para decisiones empresariales.  

**Responsabilidades:**  
- Responder preguntas de negocio con datos.  
- Crear dashboards y reportes (Tableau, Power BI).  
- Analizar tendencias y KPIs.  
- Comunicar resultados a equipos no técnicos.  

**Perfil:**  
- **Habilidades:** SQL, herramientas BI, estadística básica, narrativa.  
- **Enfoque:** "¿Qué pasó y por qué?"  

**Ejemplo:**  
Un analista en Spotify descubre que los usuarios abandonan podcasts largos en móviles. Crea un dashboard y recomienda formatos más cortos. **Impacto:** +10% en retención.  

## 3. Data Scientist: El predictor del futuro

**Misión:** Usar estadística y machine learning para **predecir y optimizar**.  

**Responsabilidades:**  
- Construir modelos predictivos (churn, recomendaciones).  
- Diseñar experimentos (A/B testing).  
- Descubrir patrones complejos.  
- Validar hipótesis con rigor científico.  

**Perfil:**    
- **Habilidades:** Python, machine learning, estadística avanzada.  
- **Enfoque:** "¿Qué pasará y qué hacemos?"  

**Ejemplo:**  
Un científico en Netflix entrena un modelo que usa historial de visualización y predice cancelaciones con 85% de precisión, identificando usuarios en riesgo. **Impacto:** Reduce churn en 3%, ahorrando millones.  

## 4. Data Engineer: El arquitecto de datos

**Misión:** Construir y mantener **pipelines de datos** confiables y escalables.  

**Responsabilidades:**  
- Diseñar ETL/ELT para mover datos.  
- Optimizar bases de datos (Snowflake, BigQuery).  
- Garantizar calidad y disponibilidad.  
- Automatizar procesos.  

**Perfil:**  
- **Formación:** Ingeniería Informática o similar.  
- **Habilidades:** Python/Scala, Spark, Airflow, cloud (AWS, GCP).  
- **Enfoque:** "¿Cómo llegan los datos rápido y limpios?"  

**Ejemplo:**  
Un ingeniero en Uber crea un pipeline que procesa 100 TB de datos de viajes en tiempo real. **Impacto:** Analistas acceden a datos frescos en minutos, no horas.  

## 5. Comparación rápida

| Aspecto | Data Analyst | Data Scientist | Data Engineer |
|---------|--------------|----------------|---------------|
| **Pregunta** | ¿Qué pasó? | ¿Qué pasará? | ¿Cómo llegan los datos? |
| **Output** | Dashboards, insights | Modelos, predicciones | Pipelines, datos limpios |
| **Herramienta clave** | SQL, Power BI | Python, Scikit-learn | Spark, Airflow |
| **Formación** | Grado general | Grado STEM + máster | Ingeniería |
| **Matemáticas** | Básica | Avanzada | Mínima |
| **Barrera de entrada** | Media | Alta | Alta |

## 6. Colaboración: El flujo de datos

**Ejemplo en Amazon:**  
1. **Ingeniero:** Construye pipeline que une datos de ventas y logística.  
2. **Analista:** Crea dashboard que muestra exceso de stock por región.  
3. **Científico:** Desarrolla modelo para predecir demanda.  
4. **Ingeniero:** Pone el modelo en producción.  
5. **Analista:** Monitoriza KPIs para medir impacto.  

**Resultado:** -15% en costes de inventario, +10% en disponibilidad.  

> **Clave:** Los ingenieros habilitan, los analistas clarifican, los científicos predicen.

## 7. ¿Cuál es tu camino?

- **Data Analyst:** Si te gusta contar historias con datos y resolver problemas de negocio.  
- **Data Scientist:** Si amas matemáticas, experimentos y programar modelos.  
- **Data Engineer:** Si disfrutas construir sistemas robustos y escalables.  

## 8. Resumen

- **Analista:** Usa SQL y BI para responder preguntas del pasado.  
- **Científico:** Usa ML para predecir el futuro.  
- **Ingeniero:** Construye el sistema que hace posible todo.  

Juntos, estos roles convierten datos crudos en **decisiones inteligentes**, formando el núcleo de un equipo de datos moderno.
