# Lección 1.2.3: Aspectos éticos y legales: GDPR y privacidad

## 1. ¿Por qué la ética y la privacidad importan en los datos?

El poder de los datos trae consigo **grandes responsabilidades**. Cada dato personal representa a una persona real, y su mal uso puede causar daños, pérdida de confianza o sanciones legales.

> **Idea clave:** No basta con lo que *podemos* hacer con los datos, sino con lo que *debemos* hacer.

**Ejemplo:**
Un modelo de recomendación que muestra ofertas médicas personalizadas podría revelar enfermedades sensibles. Ético o no depende de **cómo se usan y protegen** esos datos.

---

## 2. Principios éticos del uso de datos

| Principio                        | Descripción                                                                  | Ejemplo                                                   |
| -------------------------------- | ---------------------------------------------------------------------------- | --------------------------------------------------------- |
| **Transparencia**                | Los usuarios deben saber cómo y para qué se usan sus datos.                  | Política de privacidad clara y accesible.                 |
| **Consentimiento informado**     | Ningún dato personal debe procesarse sin permiso explícito.                  | Checkbox de aceptación al registrarse.                    |
| **Minimización**                 | Solo recolectar los datos necesarios.                                        | No pedir fecha de nacimiento si no es relevante.          |
| **Justicia y no discriminación** | Evitar sesgos que afecten a grupos vulnerables.                              | Revisar modelos de IA que favorezcan a un género o etnia. |
| **Responsabilidad**              | Las empresas deben poder explicar y justificar sus decisiones automatizadas. | Documentar modelos y sus fuentes de datos.                |

> ⚖️ **Tip:** Si no puedes explicar cómo tu algoritmo llegó a una decisión, **no lo uses en producción**.

## 3. ¿Qué es el GDPR?

El **Reglamento General de Protección de Datos (GDPR)** es una ley europea (vigente desde 2018) que protege la **privacidad de los ciudadanos** y regula cómo las empresas deben manejar sus datos personales.

### Objetivos principales:

* Dar **control total** al usuario sobre sus datos.
* Garantizar **transparencia, seguridad y consentimiento**.
* Imponer **multas severas** por incumplimiento (hasta el 4% de la facturación anual).

---

## 4. Derechos de los usuarios bajo el GDPR

| Derecho                             | Descripción                                        | Ejemplo práctico                              |
| ----------------------------------- | -------------------------------------------------- | --------------------------------------------- |
| **Acceso**                          | Saber qué datos tiene una empresa sobre ti.        | Solicitar tus datos a Facebook.               |
| **Rectificación**                   | Corregir datos inexactos.                          | Actualizar tu dirección o correo.             |
| **Supresión (“derecho al olvido”)** | Pedir que se eliminen tus datos.                   | Borrar tu cuenta y su historial.              |
| **Portabilidad**                    | Recibir tus datos en formato legible.              | Exportar tus playlists de Spotify.            |
| **Oposición**                       | Rechazar que tus datos se usen para ciertos fines. | Negarte a recibir publicidad personalizada.   |
| **Limitación del tratamiento**      | Restringir temporalmente el uso de tus datos.      | Mientras se revisa una queja de uso indebido. |

> 🔐 **Dato curioso:** El GDPR inspiró leyes similares en todo el mundo, como la CCPA en California o la LGPD en Brasil.

---

## 5. Impacto del GDPR en las organizaciones

| Área afectada           | Cambios requeridos                             | Herramientas o medidas             |
| ----------------------- | ---------------------------------------------- | ---------------------------------- |
| **Marketing**           | Obtener consentimiento antes de enviar emails. | Formularios con doble opt-in.      |
| **Analítica**           | Anonimizar datos personales en dashboards.     | Hash o pseudonimización de IDs.    |
| **Infraestructura**     | Asegurar datos en tránsito y reposo.           | Cifrado, VPN, firewalls.           |
| **Gobernanza de datos** | Documentar fuentes, accesos y flujos.          | Data Catalog, Lineage, auditorías. |

> ✅ Cumplir con GDPR no solo evita sanciones: **genera confianza** y mejora la reputación de la marca.

## 6. Casos reales

### 🏦 **Caso 1: British Airways (2019)**

* **Problema:** Filtración de datos personales y financieros de 400.000 clientes.
* **Sanción:** £183 millones (luego reducida).
* **Lección:** El descuido técnico también es un fallo de cumplimiento.

---

### 📱 **Caso 2: WhatsApp (2021)**

* **Problema:** Falta de transparencia sobre cómo compartía datos con Facebook.
* **Sanción:** €225 millones.
* **Lección:** No informar claramente al usuario es tan grave como filtrar datos.

---

### 🏥 **Caso 3: Hospital Universitario (Noruega, 2020)**

* **Problema:** Acceso indebido a historiales médicos.
* **Solución:** Auditorías internas y control de accesos automatizado.
* **Lección:** En salud, **ética y privacidad** son inseparables.

---

## 7. Checklist para cumplir ética y privacidad

* [ ] Solicita **consentimiento explícito** y claro.
* [ ] **Minimiza** los datos recolectados.
* [ ] **Anonimiza o cifra** la información sensible.
* [ ] Documenta **quién accede a qué**.
* [ ] Ofrece al usuario **control total** sobre sus datos.
* [ ] Revisa **sesgos** en modelos predictivos.
* [ ] Designa un **Data Protection Officer (DPO)** si tratas datos sensibles.

> 🧭 La ética en datos no es solo cumplir la ley: es **hacer lo correcto incluso cuando nadie mira.**

---

## 8. Resumen

* El **GDPR** protege la privacidad y la confianza de los usuarios.
* Cumplirlo requiere **transparencia, seguridad y responsabilidad**.
* La **ética de datos** debe integrarse en cada fase: desde la recolección hasta el análisis.
* Ser ético con los datos no solo evita multas: **construye relaciones de confianza duraderas**.
