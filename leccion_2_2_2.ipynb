## 3. Dataset de prÃ¡ctica: Retail Analytics

Vamos a crear un dataset realista de ventas para usar **en toda la lecciÃ³n**:

```python
np.random.seed(42)
n = 500

# Simular datos de una cadena retail
datos_ventas = pd.DataFrame({
    'mes': np.random.choice(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'], n),
    'categoria': np.random.choice(['ElectrÃ³nica', 'Ropa', 'Hogar', 'Deportes'], n),
    'ventas': np.random.normal(1000, 300, n),
    'clientes': np.random.poisson(50, n),
    'rating': np.random.uniform(1, 5, n),
    'promocion': np.random.choice(['SÃ­', 'No'], n, p=[0.3, 0.7])
})

# AÃ±adir outliers realistas (ventas excepcionales)
datos_ventas.loc[::50, 'ventas'] *= 2.5

print(datos_ventas.head())
print(f"\nDataset: {len(datos_ventas)} registros")
```

## 4. GrÃ¡ficos de distribuciÃ³n: Entender los datos

### 4.1 Histograma con KDE

**Pregunta:** Â¿CÃ³mo se distribuyen nuestras ventas?

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=datos_ventas, x='ventas', kde=True, bins=25, color='#2E86AB')
plt.title('DistribuciÃ³n de Ventas - Histograma con Curva de Densidad', fontsize=16, fontweight='bold')
plt.xlabel('Ventas (â‚¬)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.axvline(datos_ventas['ventas'].mean(), color='red', linestyle='--', linewidth=2,
            label=f"Media: {datos_ventas['ventas'].mean():.0f}â‚¬")
plt.legend()
plt.tight_layout()
plt.show()
```

**Â¿QuÃ© nos dice?**
- **Forma:** Â¿Normal, sesgada, bimodal?
- **Centro:** Media y mediana
- **DispersiÃ³n:** Â¿Muy concentrados o dispersos?
- **Outliers:** Valores extremos visibles en la cola

**Ventaja vs Matplotlib:** El parÃ¡metro `kde=True` aÃ±ade la curva de densidad automÃ¡ticamente. En Matplotlib necesitarÃ­as scipy y 10 lÃ­neas extra.

### 4.2 KDE Plot: Comparar distribuciones

**Pregunta:** Â¿Las ventas con promociÃ³n son diferentes?

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=datos_ventas, x='ventas', hue='promocion', fill=True, alpha=0.5, linewidth=2)
plt.title('ComparaciÃ³n de Ventas: Con vs Sin PromociÃ³n', fontsize=16, fontweight='bold')
plt.xlabel('Ventas (â‚¬)', fontsize=12)
plt.ylabel('Densidad', fontsize=12)
plt.tight_layout()
plt.show()
```

**Insight:** Si la curva "SÃ­" estÃ¡ desplazada a la derecha, las promociones funcionan.

## 5. Boxplot: Detectar outliers y comparar

**Pregunta:** Â¿QuÃ© categorÃ­a tiene mÃ¡s variabilidad?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos_ventas, x='categoria', y='ventas', palette='Set2', linewidth=2)
plt.title('DistribuciÃ³n de Ventas por CategorÃ­a', fontsize=16, fontweight='bold')
plt.ylabel('Ventas (â‚¬)', fontsize=12)
plt.xlabel('CategorÃ­a', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**InterpretaciÃ³n del boxplot:**
```
         â•­â”€ MÃ¡ximo (dentro de 1.5Ã—IQR)
    â”€â”€â”€â”€â”€â”¤
    â”‚    â”‚
Q3  â”œâ”€â”€â”€â”€â”¤  â† 75% de los datos estÃ¡n aquÃ­ o abajo
    â”‚    â”‚
Q2  â”œâ”€â”€â”€â”€â”¤  â† Mediana (50%)
    â”‚    â”‚
Q1  â”œâ”€â”€â”€â”€â”¤  â† 25% de los datos
    â”‚    â”‚
    â”€â”€â”€â”€â”€â”¤
         â•°â”€ MÃ­nimo
    
    â—    â† Outliers (fuera de 1.5Ã—IQR)
```

**Elementos:**
- **LÃ­nea central:** Mediana (Q2)
- **Caja:** Rango intercuartÃ­lico IQR (Q1-Q3) - donde estÃ¡ el 50% de los datos
- **Bigotes:** Valores dentro de 1.5Ã—IQR desde Q1/Q3
- **Puntos aislados:** Outliers potenciales

**Insight:** Si ElectrÃ³nica tiene mÃ¡s outliers superiores, hay productos premium que explotar.

## 6. Violin Plot: Box + DistribuciÃ³n

**Pregunta:** Â¿CÃ³mo varÃ­a la distribuciÃ³n entre categorÃ­as y promociones?

## 7. GrÃ¡ficos de relaciÃ³n: Descubrir patrones

### 7.1 Scatter Plot con regresiÃ³n

**Pregunta:** Â¿MÃ¡s clientes significan mÃ¡s ventas?

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=datos_ventas, x='clientes', y='ventas',
            scatter_kws={'alpha':0.4, 's':50, 'color':'#2E86AB'},
            line_kws={'color':'red', 'linewidth':2.5})
plt.title('RelaciÃ³n: Clientes vs Ventas (con regresiÃ³n)', fontsize=16, fontweight='bold')
plt.xlabel('NÃºmero de Clientes', fontsize=12)
plt.ylabel('Ventas (â‚¬)', fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Ventaja mÃ¡gica:** `regplot` calcula y dibuja la **lÃ­nea de regresiÃ³n** con intervalo de confianza automÃ¡ticamente. En Matplotlib necesitarÃ­as `np.polyfit` y cÃ³digo adicional.

**Insight:** Si la pendiente es pronunciada, cada cliente adicional impacta mucho en ventas.

### 7.2 Scatter multidimensional

**Pregunta:** Â¿CÃ³mo se relacionan ventas, clientes y rating por categorÃ­a?

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=datos_ventas, x='clientes', y='ventas',
                hue='categoria', size='rating', sizes=(50, 300),
                alpha=0.6, palette='deep')
plt.title('AnÃ¡lisis Multidimensional: Ventas, Clientes, Rating y CategorÃ­a', fontsize=16, fontweight='bold')
plt.xlabel('NÃºmero de Clientes', fontsize=12)
plt.ylabel('Ventas (â‚¬)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', frameon=True)
plt.tight_layout()
plt.show()
```

## 8. Pairplot: EDA explosivo

**Pregunta:** Â¿CÃ³mo se relacionan TODAS las variables entre sÃ­?

```python
# Seleccionar variables numÃ©ricas
variables_numericas = datos_ventas[['ventas', 'clientes', 'rating']]

# MAGIA: Matriz completa de relaciones
sns.pairplot(variables_numericas, diag_kind='kde', corner=False, height=3)
plt.suptitle('Pairplot - Relaciones MÃºltiples', y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Â¿QuÃ© hace `pairplot`?**
- Crea una **matriz de grÃ¡ficos**: cada variable vs cada variable.
- **Diagonal:** DistribuciÃ³n de cada variable (histograma o KDE).
- **Fuera de diagonal:** Scatter plots entre pares.
- **Todo automÃ¡tico:** 9 grÃ¡ficos en 1 lÃ­nea de cÃ³digo.

**Con categorÃ­as:**
```python
# VersiÃ³n avanzada: aÃ±adir color por categorÃ­a
sns.pairplot(datos_ventas[['ventas', 'clientes', 'rating', 'promocion']],
             hue='promocion', palette='Set1', diag_kind='kde', height=3)
plt.suptitle('Pairplot por PromociÃ³n', y=1.02, fontsize=16, fontweight='bold')
plt.show()
```

**Resultado:** En 10 segundos ves todas las relaciones posibles. En Matplotlib necesitarÃ­as 100+ lÃ­neas.

**Insight:** Si ventas y clientes se correlacionan fuerte, pero rating no, el rating no predice ventas (revisar sistema de reviews).

## 9. Heatmap: Matriz de correlaciÃ³n

**Pregunta:** Â¿QuÃ© variables estÃ¡n correlacionadas?

```python
# Calcular correlaciones
correlacion = datos_ventas[['ventas', 'clientes', 'rating']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlacion,
            annot=True,           # Mostrar valores
            fmt='.2f',            # 2 decimales
            cmap='coolwarm',      # Azul (negativo) a Rojo (positivo)
            center=0,             # Centrar en 0
            square=True,          # Celdas cuadradas
            linewidths=1.5,       # LÃ­neas entre celdas
            cbar_kws={'label': 'CorrelaciÃ³n'})
plt.title('Matriz de CorrelaciÃ³n - Variables Clave', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```

**InterpretaciÃ³n:**
- **+1.0:** CorrelaciÃ³n perfecta positiva
- **0.0:** Sin correlaciÃ³n
- **-1.0:** CorrelaciÃ³n perfecta negativa

**Insight tÃ­pico:**
- Ventas â†” Clientes: 0.65 (correlaciÃ³n moderada-alta) âœ…
- Ventas â†” Rating: 0.12 (correlaciÃ³n dÃ©bil) âš ï¸

## 13. Caso prÃ¡ctico integrado: AnÃ¡lisis completo

```python
def analisis_rapido_retail(df):
    """Dashboard completo de anÃ¡lisis retail en una funciÃ³n"""
    sns.set_theme(style="whitegrid", palette="husl")
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Dashboard Completo - Retail Analytics', fontsize=18, fontweight='bold')
    
    # 1. Variabilidad por categorÃ­a (Violin)
    sns.violinplot(data=df, x='categoria', y='ventas', ax=axes[0,0], palette='Set2')
    axes[0,0].set_title('Variabilidad de Ventas por CategorÃ­a', fontweight='bold')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # 2. Efecto promociones (Barplot con IC)
    sns.barplot(data=df, x='categoria', y='ventas', hue='promocion',
                ax=axes[0,1], estimator=np.median, errorbar='ci', palette='viridis')
    axes[0,1].set_title('Mediana de Ventas: Con/Sin PromociÃ³n', fontweight='bold')
    axes[0,1].tick_params(axis='x', rotation=45)
    axes[0,1].legend(title='PromociÃ³n')
    
    # 3. RelaciÃ³n clientes-ventas por categorÃ­a (Scatter)
    sns.scatterplot(data=df, x='clientes', y='ventas', hue='categoria',
                    alpha=0.6, s=80, ax=axes[1,0], palette='deep')
    axes[1,0].set_title('RelaciÃ³n Clientes-Ventas por CategorÃ­a', fontweight='bold')
    
    # 4. Resumen estadÃ­stico (Heatmap)
    resumen = df.groupby('categoria')['ventas'].agg(['mean', 'std', 'count'])
    sns.heatmap(resumen, annot=True, fmt='.1f', cmap='YlGnBu', ax=axes[1,1])
    axes[1,1].set_title('Resumen EstadÃ­stico por CategorÃ­a', fontweight='bold')
    
    plt.tight_layout()
    return fig

# Ejecutar anÃ¡lisis
analisis_rapido_retail(datos_ventas)
plt.show()
```

**Resultado:** Dashboard ejecutivo con 4 perspectivas en una funciÃ³n. Reutilizable para cualquier dataset retail.

**Insights tÃ­picos:**
1. **Variabilidad:** ElectrÃ³nica tiene ventas mÃ¡s dispersas (productos caros y baratos).
2. **Promociones:** Funcionan mejor en Ropa y Hogar.
3. **Clientes:** RelaciÃ³n lineal dÃ©bil, sugiere que el ticket promedio varÃ­a mucho.
4. **EstadÃ­stico:** Deportes tiene menor desviaciÃ³n estÃ¡ndar (ventas mÃ¡s predecibles).

3. **Scatterplot:** Cuenta total vs propina (con regresiÃ³n)
4. **Heatmap:** CorrelaciÃ³n entre variables numÃ©ricas
5. **Dashboard:** Combina los 4 anteriores en un subplot 2Ã—2

```python
# SoluciÃ³n sugerida
tips = sns.load_dataset('tips')

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('AnÃ¡lisis de Propinas - Restaurante', fontsize=18, fontweight='bold')

# 1. DistribuciÃ³n
sns.histplot(data=tips, x='tip', kde=True, ax=axes[0,0], color='#2E86AB')
axes[0,0].set_title('DistribuciÃ³n de Propinas')

# 2. Por dÃ­a
sns.boxplot(data=tips, x='day', y='tip', ax=axes[0,1], palette='Set2')
axes[0,1].set_title('Propinas por DÃ­a')

# 3. RelaciÃ³n
sns.regplot(data=tips, x='total_bill', y='tip', ax=axes[1,0],
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[1,0].set_title('Cuenta vs Propina')

# 4. CorrelaciÃ³n
corr = tips[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1,1])
axes[1,1].set_title('Correlaciones')

plt.tight_layout()
plt.show()
```
