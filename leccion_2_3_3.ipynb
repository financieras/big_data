# Lección 2.3.3: Looker Studio y herramientas cloud

## 1. ¿Por qué Looker Studio?

Looker Studio (antes Google Data Studio) es **el Google Docs de los dashboards**: gratuito, colaborativo y 100% en la nube.

**¿Cuándo usar Looker Studio?**  
- Necesitas dashboards **gratis** (presupuesto cero).  
- Tus datos viven en Google (Analytics, Sheets, BigQuery).  
- Quieres colaboración en tiempo real sin instalar nada.  
- Compartir con un link > enviar archivos Excel.  

> **Clave:** Si Tableau es el Ferrari y Power BI el BMW, Looker Studio es el Tesla Model 3: accesible, inteligente y perfectamente integrado en la nube.

**Dato curioso:** Google lanzó Data Studio en 2016 como respuesta gratuita a Tableau. En 2022 lo renombró "Looker Studio" tras comprar Looker por $2,600 millones. Hoy procesa más de 1 millón de reportes diarios.

## 2. Conceptos fundamentales: Cloud-first

### 2.1 ¿Qué significa "cloud-first"?

A diferencia de Tableau (desktop) o Power BI (híbrido), Looker Studio **nace en la nube**:

| Característica | Looker Studio | Tableau | Power BI |
|----------------|---------------|---------|----------|
| **¿Dónde vive?** | Navegador web | Desktop + Cloud | Desktop + Cloud |
| **Instalación** | Ninguna (0 bytes) | Sí (~2GB) | Sí (~500MB) |
| **Colaboración** | Nativa (como Docs) | Via Server/Cloud | Via Service |
| **Edición simultánea** | ✅ Sí | ❌ No | ❌ No |
| **Coste** | Gratis | €€€ (70$/mes) | € (10$/mes) |
| **Actualización** | Automática | Manual | Manual |

**La ventaja del cloud:**  
- **Colaboración:** Comparte con un link, edita en equipo.  
- **Escalabilidad:** Maneja millones de filas sin ralentizar tu PC.  
- **Actualización automática:** Los datos se refrescan solos.  
- **Acceso universal:** Solo necesitas un navegador.

### 6.3 Parámetros dinámicos (avanzado)

**Caso:** Permitir al usuario cambiar la métrica visualizada.

**Pasos:**  
1. **Recurso** → **Administrar parámetros** → **Añadir parámetro**  
2. **Nombre:** Métrica_Vista  
3. **Tipo:** Texto  
4. **Valores permitidos:** Ventas, Unidades, Beneficio, Margen  
5. **Valor predeterminado:** Ventas  

**Campo calculado dinámico:**
```javascript
CASE
  WHEN Métrica_Vista = "Ventas"
    THEN SUM(Unidades * Precio_Unit)
  WHEN Métrica_Vista = "Unidades"
    THEN SUM(Unidades)
  WHEN Métrica_Vista = "Beneficio"
    THEN SUM((Precio_Unit - Coste_Unit) * Unidades)
  WHEN Métrica_Vista = "Margen"
    THEN SUM((Precio_Unit - Coste_Unit) * Unidades) / SUM(Unidades * Precio_Unit)
END
```

**Resultado:** Un solo gráfico que cambia según la selección. Ahorra espacio y da flexibilidad máxima.

### 7.1 Campos calculados: Tu superpoder

**Sintaxis similar a Excel pero más potente:**

```javascript
// === BÁSICAS ===
Ventas_Totales = Unidades * Precio_Unit
Beneficio = (Precio_Unit - Coste_Unit) * Unidades
Margen_Porcentaje = (Beneficio / Ventas_Totales) * 100
Ticket_Promedio = SUM(Ventas_Totales) / COUNT(Pedido_ID)

// === CONDICIONALES ===
Rendimiento = CASE
  WHEN Ventas_Totales > 10000 THEN "⭐ Excelente"
  WHEN Ventas_Totales > 5000 THEN "✅ Bueno"
  WHEN Ventas_Totales > 2000 THEN "⚠️ Regular"
  ELSE "🔴 Revisar"
END

Categoria_Precio = CASE
  WHEN Precio_Unit > 1000 THEN "Premium"
  WHEN Precio_Unit > 500 THEN "Medio"
  ELSE "Económico"
END

// === FUNCIONES DE FECHA ===
Mes_Nombre = CASE MONTH(Fecha)
  WHEN 1 THEN "Enero"
  WHEN 2 THEN "Febrero"
  WHEN 3 THEN "Marzo"
  WHEN 4 THEN "Abril"
  WHEN 5 THEN "Mayo"
  WHEN 6 THEN "Junio"
  WHEN 7 THEN "Julio"
  WHEN 8 THEN "Agosto"
  WHEN 9 THEN "Septiembre"
  WHEN 10 THEN "Octubre"
  WHEN 11 THEN "Noviembre"
  WHEN 12 THEN "Diciembre"
END

Trimestre = CONCAT("Q", CAST(QUARTER(Fecha) AS TEXT))
Dia_Semana = DAYOFWEEK(Fecha)  // 1=Domingo, 7=Sábado
Es_Fin_Semana = DAYOFWEEK(Fecha) IN (1,7)

// === AGREGACIONES CONDICIONALES ===
Ventas_Online = SUM(
  CASE WHEN Canal IN ("Web", "App")
    THEN Ventas_Totales
    ELSE 0
  END
)

Ventas_Premium = SUM(
  CASE WHEN Precio_Unit > 1000
    THEN Ventas_Totales
    ELSE 0
  END
)

// === CRECIMIENTO Y COMPARACIÓN ===
Crecimiento_Abs = Ventas_Totales - Ventas_Periodo_Anterior
Crecimiento_Pct = (Crecimiento_Abs / Ventas_Periodo_Anterior) * 100

// === FUNCIONES DE TEXTO ===
Producto_Corto = LEFT(Producto, 10)
Region_Upper = UPPER(Región)
Tiene_Descuento = CONTAINS_TEXT(Campaña, "descuento")
```

> **Pro tip:** Documenta tus campos calculados en una hoja de Sheets aparte. Cuando el equipo crece, la consistencia de métricas es crítica.

### 7.2 Blending: Unir múltiples fuentes

**Caso de uso:** Combinar datos de ventas (Google Sheets) con inventario (BigQuery).

**Problema:** Quieres visualizar "Stock disponible" junto a "Unidades vendidas" para detectar roturas de stock.

**Pasos:**  
1. **Recurso** → **Administrar datos combinados**  
2. **Añadir datos combinados**  
3. **Fuente izquierda:** ventas_tech_2024 (Sheets)  
4. **Fuente derecha:** inventario_actual (BigQuery)  
5. **Clave de combinación:** Producto_ID = producto_id  
6. **Tipo de combinación:** LEFT JOIN (mantener todas las ventas)  
7. **Métricas seleccionadas:**  
   - De ventas: Unidades vendidas, Ventas_Totales  
   - De inventario: Stock_Actual, Stock_Objetivo  

**Campo calculado combinado:**
```javascript
Riesgo_Rotura = CASE
  WHEN Stock_Actual < (Unidades_Vendidas * 1.5) THEN "🔴 Alto"
  WHEN Stock_Actual < (Unidades_Vendidas * 3) THEN "🟡 Medio"
  ELSE "🟢 Bajo"
END
```

**Resultado:** Dashboard que alerta cuando un producto popular está bajo de stock.

## 9. Casos de uso reales: Empresas españolas

### Caso 1: Wallapop (Marketplace C2C)

**Problema:** No tenían visibilidad en tiempo real del rendimiento por categoría y región.  
**Solución Looker Studio:**  
- Dashboard unificado: Google Analytics + BigQuery + Sheets  
- Métricas: Conversión, productos trending, tiempo en plataforma  
- Mapa de calor: Actividad por provincia  
- Actualización automática cada 4 horas  

**Tecnología:**
```
Google Analytics 4 → BigQuery (streaming)
                  ↓
              Looker Studio
                  ↓
        Dashboard operativo 24/7
```

**Impacto:**  
- Tiempo de análisis: 2 días → 10 minutos  
- Identificaron patrones geográficos (Cataluña = tech, Andalucía = moda)  
- +18% en ventas de productos destacados mediante A/B testing  

### Caso 2: Typeform (SaaS de formularios)

**Problema:** Cada equipo (producto, ventas, marketing) usaba diferentes definiciones de métricas. Reuniones caóticas.  
**Solución Looker Studio:**  
- Single Source of Truth en BigQuery  
- Dashboard con métricas estandarizadas  
- Calculated fields documentados para MRR, Churn, LTV, CAC  
- Acceso via link para todos los equipos  

**Métricas clave estandarizadas:**
```javascript
MRR = SUM(Suscripción_Mensual)
Churn_Rate = (Cancelaciones / Clientes_Inicio_Mes) * 100
LTV = MRR / Churn_Rate
CAC = Gasto_Marketing / Nuevos_Clientes
LTV_CAC_Ratio = LTV / CAC  // Meta: >3
```

**Impacto:**  
- Reuniones 40% más eficientes  
- Decisiones basadas en datos consistentes  
- Reducción de 5 dashboards dispersos a 1 centralizado  

### Caso 3: El Corte Inglés (Retail tradicional)

**Problema:** Reportes mensuales en Excel tomaban 5 días de trabajo manual. Datos desactualizados al momento de entrega.  
**Solución Looker Studio:**  
- Migración: Excel → Google Sheets + Apps Script  
- Conectores: SAP (vía BigQuery + Cloud Functions)  
- Dashboard operativo para 90 gerentes de tienda  
- Filtros personalizados por tienda, categoría, región  

**Pipeline de datos:**
```
SAP (ventas tienda física)
        ↓
Cloud Functions (ETL cada hora)
        ↓
BigQuery (warehouse centralizado)
        ↓
Looker Studio (dashboards por perfil)
        ↓
Gerente de tienda (móvil/tablet)
```

**Impacto:**

## 10. Integración con el ecosistema cloud moderno

### 10.1 Looker Studio en un pipeline completo

**El Modern Data Stack:**
```
FUENTES (Sources)
├─ Bases de datos (MySQL, Postgres)
├─ APIs (Stripe, Shopify, Salesforce)
├─ Files (CSV, JSON, Parquet)
└─ Streaming (Kafka, Pub/Sub)
        ↓
    [EXTRACT]
        ↓
TRANSFORMACIÓN (Transform)
├─ dbt (SQL transformations)
├─ Python scripts (Pandas, PySpark)
└─ Cloud Functions (serverless)
        ↓
    [LOAD]
        ↓
DATA WAREHOUSE
├─ BigQuery (Google)
├─ Snowflake (multi-cloud)
└─ Redshift (AWS)
        ↓
    [VISUALIZE]
        ↓
LOOKER STUDIO
├─ Dashboards operativos
├─ Reportes ejecutivos
└─ Self-service BI
        ↓
    [DECISIONES]
```

**Ejemplo de flujo real (e-commerce):**
1. **Extracción:** Shopify API → Cloud Function (cada hora)  
2. **Carga:** Cloud Function → BigQuery (tabla raw)  
3. **Transformación:** dbt transforma raw → tablas analytics  
4. **Visualización:** Looker Studio conecta a BigQuery (tablas analytics)  
5. **Acción:** Equipo de marketing ajusta campañas según dashboard  

### 10.2 BigQuery + Looker Studio: La combinación perfecta

**¿Por qué BigQuery?**  
- ⚡ **Velocidad:** Consultas sobre TB en segundos  
- 💰 **Coste:** Solo pagas por datos procesados ($5 por TB)  
- 📊 **Escalabilidad:** De GB a PB sin configuración  
- 🔗 **Integración:** Nativa con Looker Studio  

**Consulta SQL optimizada para Looker:**
```sql
-- Ejemplo: Dashboard de ventas optimizado
SELECT
  DATE_TRUNC(fecha_pedido, MONTH) as mes,
  region,
  categoria_producto,
  SUM(cantidad * precio_unitario) as ventas_totales,
  SUM(cantidad) as unidades_vendidas,
  COUNT(DISTINCT cliente_id) as clientes_unicos,
  AVG(cantidad * precio_unitario) as ticket_promedio
FROM `proyecto.dataset.ventas`
WHERE fecha_pedido >= DATE_SUB(CURRENT_DATE(), INTERVAL 12 MONTH)
GROUP BY mes, region, categoria_producto
ORDER BY mes DESC, ventas_totales DESC
```

**Configuración en Looker Studio:**
1. **Crear fuente de datos** → **BigQuery**  
2. **Autorización:** Permitir acceso a proyecto  
3. **Modo:** Custom Query (pegar SQL arriba)  
4. **Caché:** Activar (12 horas) para reducir costes  
5. **Conectar:** Listo para crear visualizaciones  

> **Ahorro de costes:** Con caché activado, una consulta de 1TB cuesta $5 una vez cada 12h (vs $5 cada vez sin caché).

## 12. Errores comunes y soluciones

### 12.1 Problemas técnicos

| Error | Síntoma | Causa | Solución |
|-------|---------|-------|----------|
| **"No se pueden cargar los datos"** | Gráficos en blanco | Fuente de datos sin permisos | Compartir Data Source con usuarios |
| **"Límite de consultas excedido"** | Error en BigQuery | Demasiadas queries sin caché | Activar caché (12h) |
| **Dashboard lento** | Tarda >10 seg en cargar | Consulta ineficiente o fuente lenta | Optimizar SQL o usar extracts |
| **Números no cuadran** | Métricas inconsistentes | Múltiples definiciones de campos | Estandarizar calculated fields |
| **"Dimensión no disponible"** | Error en blend | Campo no existe en ambas fuentes | Verificar join keys |
| **Gráfico desconfigurado** | Diseño roto después de editar | Cambio en estructura de datos | Re-mapear dimensiones/métricas |

### 12.2 Errores de diseño

| Error | Consecuencia | Cómo evitarlo |
|-------|--------------|---------------|
| **Demasiados gráficos** | Parálisis por análisis | Máximo 6-8 elementos por página |
| **Sin jerarquía visual** | Usuario no sabe qué mirar primero | KPIs grandes arriba, detalles abajo |
| **Colores aleatorios** | Dashboard poco profesional | Define paleta de 3-5 colores |
| **Gráficos sin contexto** | "¿Esto es bueno o malo?" | Añade comparaciones vs período/objetivo |
| **Olvidar mobile** | Dashboard roto en teléfono | Probar en **Vista de móvil** |
| **Compartir sin probar** | Usuarios ven errores | Checklist de QA antes de compartir |

### 12.3 Checklist antes de compartir

## 14. Ejercicio práctico: Dashboard de marketing digital

**Contexto:** Eres analista de datos en "EcoTech", startup de gadgets sostenibles. El CMO quiere un dashboard para optimizar campañas.

### Dataset: marketing_campaigns_2024

Crea este Google Sheet con los siguientes datos:

| Fecha | Canal | Campaña | Impresiones | Clics | Conversiones | Coste | Ingresos |
|-------|-------|---------|-------------|-------|--------------|-------|----------|
| 2024-01-15 | Google Ads | Lanzamiento | 25,000 | 1,250 | 45 | 850 | 12,500 |
| 2024-01-16 | Meta Ads | Eco-Friendly | 18,500 | 980 | 38 | 620 | 9,800 |
| 2024-01-17 | Email | Newsletter | 15,000 | 2,100 | 120 | 150 | 15,200 |
| 2024-01-18 | TikTok | Gen Z Target | 42,000 | 3,400 | 85 | 1,200 | 8,500 |
| 2024-01-22 | Google Ads | Retargeting | 12,000 | 840 | 52 | 450 | 14,040 |
| 2024-01-23 | LinkedIn | B2B Outreach | 8,500 | 510 | 28 | 380 | 8,400 |
| 2024-01-29 | Orgánico | Blog SEO | 5,200 | 420 | 35 | 0 | 10,500 |

### Objetivos del dashboard

1. **Identificar canal con mejor ROI**  
2. **Analizar tendencia de conversiones**  
3. **Calcular CPA (Coste por Adquisición)**  
4. **Comparar eficiencia de campañas**  

### Tareas paso a paso

**1. Preparación (15 min)**
- Crea el Google Sheet con los datos arriba  
- Nombra la hoja: `marketing_campaigns_2024`  
- Ve a Looker Studio y conecta esta fuente  

**2. Campos calculados (10 min)**

Crea estos campos calculados:
```javascript
// ROI (Retorno sobre Inversión)
ROI = ((Ingresos - Coste) / Coste) * 100

// CTR (Click Through Rate)
CTR = (Clics / Impresiones) * 100

// CPA (Coste por Adquisición)
CPA = Coste / Conversiones

// Tasa de Conversión
Tasa_Conversion = (Conversiones / Clics) * 100

// Beneficio Neto
Beneficio = Ingresos - Coste
```

**3. Visualizaciones (25 min)**

Crea los siguientes gráficos:

a) **Scorecards principales** (arriba):
   - ROI promedio (con comparación vs objetivo 300%)  
   - Total conversiones  
   - Beneficio total  
   - CPA promedio  

b) **Gráfico de barras**: ROI por canal (ordenado desc)

c) **Serie temporal**: Conversiones por día con línea de tendencia

d) **Tabla con heatmap**: Canal × Campaña mostrando:
   - Conversiones  
   - CPA  
   - ROI  
   - Beneficio  

e) **Gráfico de dispersión**: CPA (eje X) vs Conversiones (eje Y)
   - Añade etiquetas con nombre de campaña  
   - Objetivo: encontrar el sweet spot (bajo CPA + altas conversiones)  

**4. Interactividad (5 min)**

Añade controles:
- Filtro de canal (lista desplegable, múltiple selección)  
- Selector de fechas (rango)  
- Filtro de campaña (búsqueda)  

**5. Diseño (5 min)**

- Aplica paleta de colores: Verde = bueno (ROI alto), Rojo = revisar (ROI bajo)  
- Añade título: "Dashboard Marketing Digital - EcoTech"  
- Incluye fecha de última actualización  
- Logo de empresa (usa placeholder si no tienes)  

**Tiempo total:** 60 minutos

### Insights esperados

Al completar el ejercicio, deberías descubrir:

✅ **Email marketing tiene mejor ROI** (9,900% - bajo coste, alta conversión)  
✅ **TikTok genera mucho tráfico pero ROI bajo** (608% - mucho coste)  
✅ **Google Ads Retargeting es muy eficiente** (CPA €8.65, ROI 3,020%)  
✅ **Orgánico es 100% rentable** (coste €0, ingresos €10,500)  

**Recomendaciones al CMO:**
1. Aumentar inversión en email marketing (ROI altísimo)  
2. Optimizar campañas de TikTok o reducir presupuesto  
3. Escalar retargeting en Google Ads  
4. Invertir en SEO para crecer el canal orgánico
