# Lección 1.1.2: Historia y evolución del ecosistema de datos

## 1. Los orígenes: el problema del volumen (1990s–2000s)

En los años 90, las empresas ya almacenaban grandes volúmenes de datos en **bases de datos relacionales** (Oracle, SQL Server) y **data warehouses** (Teradata).
El problema: **coste, lentitud y falta de escalabilidad** para internet.

Con el crecimiento de Internet, Google enfrentó un reto nuevo: indexar **miles de millones de páginas web**.
Para resolverlo, publicó dos papers clave:

* **2003:** *Google File System (GFS)* → almacenamiento distribuido
* **2004:** *MapReduce* → procesamiento paralelo

> Estos dos artículos inspiraron el nacimiento del Big Data moderno.

---

## 2. Nace Hadoop (2006): el Big Data se democratiza

**Doug Cutting** y **Mike Cafarella** crean **Hadoop**, una versión *open source* de las ideas de Google.
Por primera vez, cualquier empresa podía procesar datos a gran escala usando **clusters de servidores comunes**.

* Yahoo! y Facebook fueron los primeros en adoptarlo.
* Surgen herramientas herramientas como **Hive** (SQL-like), **Pig** (scripting) y **HBase** (bases NoSQL).

👉 Hadoop resolvía el volumen, pero era **complejo y lento** y solo para procesamiento por lotes (*batch*)..

---

## 3. Llega Spark (2014): velocidad y simplicidad

**Apache Spark** revolucionó el procesamiento de datos:

* 100× más rápido que Hadoop (usa memoria RAM).
* Permite **batch y streaming**.
* Se volvió el nuevo estándar de Big Data.

**Ejemplo**: Netflix adoptó Spark para recomendaciones más rápidas.

> Spark fue al Big Data lo que los motores eléctricos fueron a la industria: hizo todo más rápido y eficiente.

---

## 4. La nube: escalabilidad para todos (2015–2018)

La nube (AWS, Google Cloud, Azure) eliminó la necesidad de clusters físicos:
- Servicios gestionados como Amazon S3, BigQuery o Azure Data Lake.
- Modelo serverless: paga solo por lo que usas.
- Democratización: Las Pymes accedieron a tecnología de gigantes como Google o Netflix.

**Ejemplo**: Una startup en 2016 podía analizar terabytes sin invertir en servidores.

---

## 5. Del Big Data al Ecosistema Moderno (2018–presente)

El foco cambió de **“cuántos datos tenemos”** a **“qué valor extraemos de ellos”**.

El Big Data evolucionó al Modern Data Stack:
- **Foco**: Extraer valor, no solo procesar datos.
- **Herramientas modulares**: Cada una hace una cosa bien (Fivetran para ingesta, Snowflake para almacenamiento, dbt para transformación, Looker para visualización).
- **Cloud-native**: Sin infraestructura propia.
- **SQL renace**: Analistas trabajan sin programar en Spark.
- **Data Lakehouse**: Combina lo mejor de Data Lakes y Warehouses (Databricks, Snowflake).

### Cambios clave:

| Antes (2010–2018) | Ahora (2025) |
|-------------------|--------------|
| Clusters propios | Cloud/serverless |
| ETL (transformar antes) | ELT (transformar en destino) |
| Procesamiento batch | Tiempo real + streaming |
| Data Lake                  | **Data Lakehouse**                                         |
| Data Scientist “hace todo” | Roles especializados (Data Engineer, Analyst, ML Engineer) |

> Hoy el objetivo no es solo procesar datos, sino **convertirlos en decisiones y productos inteligentes**.

---

## 6. Tendencias clave en la actualidad

- **Data Lakehouses:** Formatos abiertos (Delta Lake, Iceberg) para datos estructurados y no estructurados.  
- **Democratización:** Analistas usan SQL, no lenguajes complejos.  
- **Calidad de datos:** Herramientas como Great Expectations aseguran fiabilidad.  
- **IA integrada:** Feature Stores y MLOps para modelos predictivos.  

**Ejemplo:** Airbnb usa un Data Lakehouse para analizar reservas y precios en tiempo real.

## 7. Resumen

- **1990s–2006:** Bases relacionales limitadas → Google crea GFS y MapReduce.  
- **2006–2014:** Hadoop democratiza el Big Data, pero es lento.  
- **2014–2018:** Spark acelera y la nube simplifica.  
- **2018–2025:** Modern Data Stack: modular, cloud-native, orientado al valor.  

> El Big Data no desapareció: **evolucionó** hacia un ecosistema más ágil  que transforma datos en decisiones, escalable y accesible para todos.
